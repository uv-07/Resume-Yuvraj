# Machine Learning Model

```python
df = data_frame.copy()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


X_train_scaled
```

```python
model = RandomForestClassifier(random_state=0)
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)

# Print the confusion matrix and classification report
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
y_pred.shape
```

```python
y_test.shape
```

```python
residuals = y_test - y_pred

sns.histplot(residuals, kde=True, color='blue')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Residuals')
plt.show()
```

```python

# Extract correlations with the 'Class' column
class_corr = corr_matrix['Class']

# Set a threshold for "low" correlation (e.g., |correlation| < 0.1)
threshold = 0.1
low_corr_features = class_corr[abs(class_corr) < threshold].index

# Exclude 'Class' itself


print("Columns with low correlation with 'Class':")
print(low_corr_features)
```
