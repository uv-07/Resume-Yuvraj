Q. 2 Principal Component Analysis:

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, decomposition

np.random.seed(5)

iris = datasets.load_iris()
X = iris.data
y = iris.target


X_mean = np.mean(X, axis=0)
X_centered = X - X_mean

covariance_matrix = np.cov(X_centered, rowvar=False)

eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)


k = 3
top_eigenvectors = eigenvectors[:, :k]

reduced_data = np.dot(X_centered, top_eigenvectors)

# print(reduced_data)

fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = fig.add_subplot(111, projection="3d", elev=48, azim=134)
ax.set_position([0, 0, 0.95, 1])
for name, label in [("Setosa", 0), ("Versicolour", 1), ("Virginica", 2)]:
    ax.text3D(
        reduced_data[y == label, 0].mean(),
        reduced_data[y == label, 1].mean() + 1.5,
        reduced_data[y == label, 2].mean() ,
        name,
        horizontalalignment="center",
        bbox=dict(alpha=0.5, edgecolor="w", facecolor="w"),
    )
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(float)
ax.scatter(reduced_data[:, 0], reduced_data[:, 1], reduced_data[:, 2], c=y, cmap=plt.cm.nipy_spectral, edgecolor="k")

ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

plt.show()
```
